```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

## Distribuição Normal

```python
# Uma cesta de objetos, a média é 8 o desvio padrão é 2, qual a probabilidade de de tirar um objeto < 6 quilos
# cdf para valores menores < que
norm.cdf(6, 8, 2)
```

```python
# sf para valores maior > que
norm.sf(6, 8, 2)
```

```python
1 - norm.cdf(6, 8, 2)
```

```python
# Qual a probabilidade de tirar um objeto < 6 e > 10
norm.cdf(6, 8, 2) + norm.sf(10, 8, 2)
```

```python
# Qual a probabilidade de tirar um objeto < 10 e > 8
norm.cdf(10, 8, 2) - norm.cdf(8, 8, 2)
```

## Verificando Normalidade

```python
# Gerando uma amostra de tamnho 100 normal
dados = norm.rvs(size = 1000)
dados
```

```python
# Histograma
plt.hist(dados, bins = 20)
plt.title('Dados')
```

```python
fig, ax = plt.subplots()
stats.probplot(dados, fit = True, plot = ax)
plt.show()
```

```python
# Realizando o teste de Shapiro
stats.shapiro(dados)
```

```python
# Dados não normais
dados2 = skewnorm.rvs(4, size = 100)
dados2
```

```python
# Histograma
plt.hist(dados2, bins = 20)
plt.title('Dados2')
plt.show()
```

```python
fig, ax = plt.subplots()
stats.probplot(dados2, fit = True, plot = ax)
plt.show()
```

```python
# Realizando o teste de Shapiro
stats.shapiro(dados)
```
